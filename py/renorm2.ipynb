```python
Hp = 2
dirs = dhnnlib.MAKE_DIRTREE(Hp=Hp)
avgm = 10
ptyp = "MARKOV"
infn = f"{dirs.DINPT}renorm_in{etxt}"
H = dhnnlib(#
    Hp=Hp, 
    infn=infn,
    ptyp=ptyp,
    avgm=avgm,
    AVGD=True,
    DATAF=dirs.D4PLT
)
```

```python
Fdic = [H.MAKE__KTiXN_J(N) for N in H.lsN]
vs = Fdic[0]
N = vs['N']
NHp = N**(H.Hp-1)
lsK = [K for K in sorted(H.GET__UvXk(vs['TiK'], 'K'))]
lsKTi = np.unique([[K, Ti] for K in lsK
                 for Ti in H.MAKE__T_iperK(Fdic[0], N, 0, K)], axis=0)
lsKTi = np.array([tuple(row) for row in lsKTi], dtKTi)
```

```python
lsEig = []
for i in lsKTi:
    J = H.MAKE__Jmat(N, *i)/N
    eig = H.MAKE__JLeigSM(np.linalg.matrix_power(J, 2))
    lsEig.append([*i, eig])
```
